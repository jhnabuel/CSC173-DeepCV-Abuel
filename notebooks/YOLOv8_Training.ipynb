# Training YOLOv8 small model (base model)

```python
from ultralytics import YOLO

# Load Base Model (Small version, matching research)
model = YOLO('yolov8s.pt')

# Train
results = model.train(
    data='/content/dataset/yolo_official/data.yaml',
    epochs=100,
    imgsz=512,              # Dataset standard
    batch=16,
    patience=15,            # Early stopping if no improvement
    name='tbx11k_base_model',
    project='/content/drive/MyDrive/TB_Project/runs'
)
```
